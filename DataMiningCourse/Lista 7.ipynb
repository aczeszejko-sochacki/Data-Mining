```python
mean1, mean2, mean3 = [-21, -2], [3, 5], [27, 12]
cov = [[12, 3], [3, 1]]
x = []
y = []
for i in range(1000):
    which = np.random.choice([1, 2, 3])
    if which == 1:
        rand = np.random.multivariate_normal(mean1, cov, 1).T
        x.append(rand[0][0])
        y.append(rand[1][0])
    if which == 2:
        rand = np.random.multivariate_normal(mean2, cov, 1).T
        x.append(rand[0][0])
        y.append(rand[1][0])
    if which == 3:
        rand = np.random.multivariate_normal(mean3, cov, 1).T
        x.append(rand[0][0])
        y.append(rand[1][0])

x = np.array(x)
y = np.array(y)
X = np.array([x, y])
plt.scatter(x, y, s=10)
plt.show()
```

```python
eig = LA.eig(S)
λ1 = max(eig[0][0], eig[0][1])
λ2 = min(eig[0][0], eig[0][1])
v1 = eig[1][0]
v2 = eig[1][1]
print('Eigenvalues:\n' + str(eig[0]))
print('Eigenvectors:\n' + str(eig[1][0]) + "\n" + str(eig[1][1]))
```

## j)

```python
iris = datasets.load_iris().data
```

```python
X_0 = preprocessing.scale(iris)
```

```python
S = np.cov(X_0.T)
S
```

```python
eig = LA.eig(S)
v1 = eig[1][0]
v2 = eig[1][1]
print('Eigenvalues:\n' + str(eig[0][0]) + "\n" + str(eig[0][1]) + "\n" + str(eig[0][2]) + "\n" + str(eig[0][3]))
print('Eigenvectors:\n' + str(eig[1][0]) + "\n" + str(eig[1][1]) + "\n" + str(eig[1][2]) + "\n" + str(eig[1][3]))
```

```python
points = list(zip(X_0.T[0], X_0.T[1], X_0.T[2], X_0.T[3]))
def pca_scale(point):
    return [1 / sqrt(λ1) * np.dot(v1, point), 1 / sqrt(λ2) * np.dot(v2.T, point)]


λ1, λ2 = eig[0][0], eig[0][1]
print(λ1, λ2)
PCA = np.array(list(zip(*list(map(pca_scale, points)))))
plt.scatter(PCA[0][:50], PCA[1][:50])
plt.scatter(PCA[0][51:100], PCA[1][51:100])
plt.scatter(PCA[0][100:150], PCA[1][100:150])
plt.show()
```

```python
pca_to_rec = list(map(pca_scale, points))
reconstructed = np.dot(pca_to_rec, np.array([v1, v2]))
reconstructed
```

```python
mean_squared_error(iris, reconstructed)
```

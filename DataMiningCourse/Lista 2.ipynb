## zadanie 1

```python
from sklearn import datasets
import matplotlib.pyplot as plt

iris = datasets.load_iris()

l = len(iris.target)

sepal_length_setosa = [iris.data[i][0] for i in range(l) if iris.target[i] == 0]
sepal_width_setosa = [iris.data[i][1] for i in range(l) if iris.target[i] == 0]
sepal_length_veriscolor = [iris.data[i][0] for i in range(l) if iris.target[i] == 1]
sepal_width_versicolor = [iris.data[i][1] for i in range(l) if iris.target[i] == 1]
sepal_length_virginica = [iris.data[i][0] for i in range(l) if iris.target[i] == 2]
sepal_width_virginica = [iris.data[i][1] for i in range(l) if iris.target[i] == 2]

plt.scatter(x=sepal_length_setosa, y=sepal_width_setosa, color='red')
plt.scatter(x=sepal_length_veriscolor, y=sepal_width_versicolor, color='green')
plt.scatter(x=sepal_length_virginica, y=sepal_width_virginica, color='blue')

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.xticks(range(3, 10))
plt.yticks(range(1, 6))
plt.title('Iris')
plt.savefig('zadanie1.png')
plt.show()
```

## zadanie 2

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np

iris = datasets.load_iris()

l = len(iris.target)

sepal_length_setosa_good = [iris.data[i][0] for i in range(l) if iris.target[i] == 0
                       and iris.data[i][0]*2 - 8 < iris.data[i][1]]
sepal_length_setosa_bad = [iris.data[i][0] for i in range(l) if iris.target[i] == 0
                       and iris.data[i][0]*2 - 8 >= iris.data[i][1]]

sepal_width_setosa_good = [iris.data[i][1] for i in range(l) if iris.target[i] == 0
                           and iris.data[i][1]/2 + 4 > iris.data[i][0]]
sepal_width_setosa_bad = [iris.data[i][1] for i in range(l) if iris.target[i]
                          and iris.data[i][0]/2 + 4 <= iris.data[i][0]]

sepal_length_virginica_good = [iris.data[i][0] for i in range(l) if iris.target[i] == 2
                       and iris.data[i][0]*2 - 8 > iris.data[i][1]]
sepal_length_virginica_bad = [iris.data[i][0] for i in range(l) if iris.target[i] == 2
                       and iris.data[i][0]*2 - 8 <= iris.data[i][1]]

sepal_width_virginica_good = [iris.data[i][1] for i in range(l) if iris.target[i] == 2
                              and iris.data[i][1]/2 + 4 < iris.data[i][0]]
sepal_width_virginica_bad = [iris.data[i][1] for i in range(l) if iris.target[i] == 2
                             and iris.data[i][1]/2 + 4 > iris.data[i][0]]

sepal_length_good = sepal_length_setosa_good + sepal_length_virginica_good
sepal_width_good = sepal_width_setosa_good + sepal_width_virginica_good
sepal_length_bad = sepal_length_setosa_bad + sepal_length_virginica_bad
sepal_width_bad = sepal_width_setosa_bad + sepal_width_virginica_bad

func_x = np.array(list(range(4, 8)))
func_y = func_x*2 - 8
func_better = func_x*1.4 - 4.3


plt.plot(func_x, func_y)
plt.plot(func_x, func_better)
plt.scatter(x=sepal_length_good, y=sepal_width_good, color='green')
plt.scatter(x=sepal_length_bad, y=sepal_width_bad, color='red')

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.xticks(range(3, 10))
plt.yticks(range(1, 6))
plt.title('Iris')

plt.show()
```

## zadanie 3

```python
import matplotlib.pyplot as plt
import numpy as np
import math

x_points = np.array(list(range(1, 11)))
y_points = [10, 10, 11, 12, 18, 18, 19, 26, 19, 26]


def dist(factor, value):
    distance = [math.fabs(y_points[i] - (factor*x_points[i] + value)) for i in range(10)]

    print(distance)
    print(sum(distance))
    return factor*x_points + value


plt.scatter(x_points, y_points, color='red')
plt.plot(x_points, dist(2, 5), color='purple')
plt.savefig('zadanie3a.png')
plt.plot(x_points, dist(2, 6), color='magenta')
plt.savefig('zadanie3b.png')
plt.show()
```

## zadanie 4

```python
import matplotlib.pyplot as plt
import numpy as np
import math


def centers(k, d, sigma):
    for angle in range(k):
        x_points = np.random.normal(d * math.sin(angle/k * 2*math.pi), sigma, 1000)
        y_points = np.random.normal(d * math.cos(angle/k * 2*math.pi), sigma, 1000)
        plt.scatter(x_points, y_points, alpha=0.5, s=4)

    plt.savefig("zadanie4_" + str(k) + "_" + str(d) + ".png")
    plt.show()


centers(7, 5, 1)
centers(11, 10, 1)
centers(23, 15, 1)
```
